# Heat map:

- data visualization technique used to display the intensity of values in a matrix format
- It is commonly used to visualize correlations or data distributions in a two-dimensional dataset.
  
Important Parameter:

- data: The data to be displayed, usually a 2D array or pandas DataFrame
- vmin, vmax: Minimum and maximum values for the colormap. Useful for normalizing the scale
- annot: If True, displays the data values in each cell
- annot_kws: Dictionary for customizing annotations (e.g., {"size": 10, "color": "black"})
- fmt: String formatting code for annotations (e.g., "d" for integers, ".2f" for floating-point numbers)
- cmap: Specifies the colormap (e.g., "viridis", "coolwarm", "YlGnBu")
- center: Value at which to center the colormap (useful for divergent colormaps)
- cbar: If True, includes a color bar.
- linewidths: Width of the lines between cells.
- linecolor: Color of the lines between cells.

```python
glue=sns.load_dataset("glue")
glue.head()
```

```python
glue=sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
sns.heatmap(glue)
```
